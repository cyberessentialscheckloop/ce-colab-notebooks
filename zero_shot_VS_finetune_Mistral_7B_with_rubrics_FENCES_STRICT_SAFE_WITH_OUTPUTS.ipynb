### Zero shot model

```python
# ────────────────
# Zero-Shot Rubric-Aware Inference for Mistral-7B-Instruct
# ────────────────

import pandas as pd
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,
    ConfusionMatrixDisplay, classification_report
)
from pathlib import Path
import re
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from rubrics import RUBRICS as rubrics  # <-- your rubric file

# 1. Load labelled data (labels never shown to model!)
CSV_PATH = "/content/drive/MyDrive/project/CB/balanced_undersampled_per_bucket.csv"
df = pd.read_csv(CSV_PATH)
rows = df[["question_id", "question_text", "answer_text"]].to_dict("records")
ground_truth = df["label"].str.title().tolist()  # Only for metrics

# 2. Load BASE model and tokenizer (no LoRA/PEFT, no fine-tuned weights)
MODEL_ID = "mistralai/Mistral-7B-Instruct-v0.2"
base_model = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float16,
    device_map="auto",
    trust_remote_code=True,
    low_cpu_mem_usage=True,
)
tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True)
tokenizer.pad_token = tokenizer.eos_token

# 3. Prompt builder (labels/reasons NEVER in prompt!)
SYSTEM_MSG = (
    "You are a Cyber Essentials assessor.\n"
    "Classify the applicant answer as Compliant or Non-Compliant, using the rubric below.\n"
    "Reply with ONLY TWO lines:\n"
    "LABEL: <Compliant or Non-Compliant>\n"
    "REASON: <Short reason, quoting the rubric criteria met or missed.>"
)

def build_prompt(row):
    rubric = rubrics.get(row["question_id"])
    criteria = "\n".join(rubric["criteria"]) if rubric else "No rubric found."
    user_msg = (
        f"Question:\n{row['question_text']}\n\n"
        f"Applicant answer:\n{row['answer_text']}\n\n"
        f"Rubric:\n{criteria}"
    )
    return (
        "<s>[INST] <<SYS>>\n"
        f"{SYSTEM_MSG}\n"
        "<</SYS>>\n"
        f"{user_msg}\n"
        "[/INST]"
    )

# 4. Output extraction (LABEL only)
label_re = re.compile(r"(?i)label\s*:\s*(compliant|non[\s-]?compliant)")
def extract_label(text):
    m = label_re.search(text)
    if not m: return None
    raw = m.group(1).lower().replace(" ", "-")
    return "Compliant" if raw.startswith("compliant") else "Non-Compliant"

# 5. Batch inference loop
BATCH_SIZE = 6
prompts = [build_prompt(row) for row in rows]
preds = []

for i in tqdm(range(0, len(prompts), BATCH_SIZE), desc="Zero-shot Inference"):
    batch_prompts = prompts[i:i+BATCH_SIZE]
    inputs = tokenizer(batch_prompts, return_tensors="pt", padding=True, truncation=True, max_length=512)
    inputs = {k: v.to(base_model.device) for k, v in inputs.items()}

    with torch.no_grad():
        out = base_model.generate(
            **inputs,
            max_new_tokens=64,
            do_sample=False,
            pad_token_id=tokenizer.eos_token_id,
            eos_token_id=tokenizer.eos_token_id,
        )
    decoded = tokenizer.batch_decode(out, skip_special_tokens=True)
    for txt in decoded:
        preds.append(extract_label(txt) or "UNKNOWN")


```

```python
# 6. Metrics calculation

LABELS = ["Compliant", "Non-Compliant", "UNKNOWN"]
cm = confusion_matrix(ground_truth, preds, labels=LABELS)

# --- 6a. Metrics EXCLUDING UNKNOWN (filtered) ---
valid_mask = [(g in ("Compliant", "Non-Compliant")) and (p in ("Compliant", "Non-Compliant")) for g, p in zip(ground_truth, preds)]
filt_gold = [g for g, v in zip(ground_truth, valid_mask) if v]
filt_pred = [p for p, v in zip(preds, valid_mask) if v]

acc  = accuracy_score(filt_gold, filt_pred)
prec = precision_score(filt_gold, filt_pred, pos_label="Compliant")
rec  = recall_score(filt_gold, filt_pred, pos_label="Compliant")
f1   = f1_score(filt_gold, filt_pred, pos_label="Compliant")

print("=== Zero-shot metrics (excluding UNKNOWN) ===")
print(f"Accuracy (excluding UNKNOWN):  {acc:.4f}")
print(f"Precision: {prec:.4f}")
print(f"Recall:    {rec:.4f}")
print(f"F1 Score:  {f1:.4f}")
print("-" * 40)

# --- 6b. Metrics INCLUDING UNKNOWN as error class ---
acc_strict = accuracy_score(ground_truth, preds)
report_strict = classification_report(ground_truth, preds, labels=LABELS, zero_division=0, digits=4)

print("=== Strict metrics (including UNKNOWN as error class) ===")
print(f"Strict Accuracy (UNKNOWN = error): {acc_strict:.4f}")
print("\nPer-class metrics (INCLUDING UNKNOWN):")
print(report_strict)
print("\nConfusion Matrix (INCLUDING UNKNOWN):")
print(pd.DataFrame(cm, index=LABELS, columns=LABELS))

n_unknown = (pd.Series(preds) == "UNKNOWN").sum()
print(f"\nTotal predictions: {len(preds)}  |  Unknowns: {n_unknown}  ({n_unknown / len(preds):.2%})")

# 7. Plot confusion matrix with UNKNOWN
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=LABELS)
fig, ax = plt.subplots(figsize=(6, 6))
disp.plot(ax=ax, cmap="Blues", colorbar=False, values_format='d')
plt.title("Confusion Matrix (including UNKNOWN)")
plt.tight_layout()
plt.show()
```

### Training in 5-cross validation folds and Evaluation

```python
# mistral_ce_assessor_fixed.py
"""
Fine‑tune Mistral‑7B‑Instruct‑v0.2 with LoRA to classify Cyber‑Essentials answers
───────────────────────────────────────────────────────────────────────────────
Key points
* Supervision format == inference format (two‑line LABEL / REASON)
* Adds BOS token "<s>" expected by Mistral‑Instruct
* Prompt tokens masked with ‑100 in labels
* Robust label extraction via regex
* Custom collator pads *labels* as well as inputs
"""

import gc
import json
import re
from pathlib import Path

import numpy as np
import pandas as pd
import torch
from datasets import Dataset
from peft import LoraConfig, get_peft_model
from sklearn.metrics import (accuracy_score, confusion_matrix, f1_score,
                             precision_score, recall_score)
from sklearn.model_selection import StratifiedKFold
from torch.nn.utils.rnn import pad_sequence
from tqdm.auto import tqdm
from transformers import (AutoModelForCausalLM, AutoTokenizer, Trainer,
                          TrainingArguments, set_seed)

from rubrics import RUBRICS as rubrics      # ← your existing file

# ──────────────────────────────────────────────────────────────────────────
# Config
# ──────────────────────────────────────────────────────────────────────────
SEED = 123
set_seed(SEED)

MODEL_ID   = "mistralai/Mistral-7B-Instruct-v0.2"
CSV_PATH   = "/content/drive/MyDrive/project/CB/balanced_undersampled_per_bucket.csv"
OUTPUT_DIR = Path("/content/drive/MyDrive/project/CB")

lora_cfg = LoraConfig(
    r=16, lora_alpha=32, lora_dropout=0.05, bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True)
tokenizer.pad_token = tokenizer.eos_token

SYSTEM_MSG = (
    "You are a Cyber Essentials assessor.\n"
    "Classify the applicant answer as Compliant or Non-Compliant, using the rubric below.\n"
    "Reply with ONLY TWO lines:\n"
    "LABEL: <Compliant or Non-Compliant>\n"
    "REASON: <Short reason, quoting the rubric criteria met or missed.>"
)

# ──────────────────────────────────────────────────────────────────────────
# Prompt helpers
# ──────────────────────────────────────────────────────────────────────────
def build_prompt(row: dict) -> str:
    rubric = rubrics.get(row["question_id"])
    criteria = "\n".join(rubric["criteria"]) if rubric else "No rubric found."
    user_msg = (
        f"Question:\n{row['question_text']}\n\n"
        f"Applicant answer:\n{row['answer_text']}\n\n"
        f"Rubric:\n{criteria}"
    )
    return (
        "<s>[INST] <<SYS>>\n"
        f"{SYSTEM_MSG}\n"
        "<</SYS>>\n"
        f"{user_msg}\n"
        "[/INST]"
    )

def build_sft_example(row: dict) -> dict:
    prompt   = build_prompt(row)
    response = f"LABEL: {row['label'].title()}\nREASON: {row.get('reason', '')}"
    return {"prompt": prompt, "response": response}

# ──────────────────────────────────────────────────────────────────────────
# Tokenisation
# ──────────────────────────────────────────────────────────────────────────
def tokenize_supervised(batch):
    p_tok = tokenizer(batch["prompt"],   truncation=True, max_length=512, add_special_tokens=False)
    r_tok = tokenizer(batch["response"], truncation=True, max_length=128, add_special_tokens=False)

    input_ids, attention_masks, labels = [], [], []
    for p_ids, r_ids in zip(p_tok["input_ids"], r_tok["input_ids"]):
        ids   = p_ids + r_ids + [tokenizer.eos_token_id]
        mask  = [1] * len(ids)
        lbls  = [-100] * len(p_ids) + r_ids + [tokenizer.eos_token_id]

        input_ids.append(ids)
        attention_masks.append(mask)
        labels.append(lbls)

    return {
        "input_ids": input_ids,
        "attention_mask": attention_masks,
        "labels": labels,
    }

# Custom padding collator that keeps ‑100 for padding in labels
def causal_lm_collator(features):
    pad_id = tokenizer.pad_token_id
    ids   = pad_sequence([torch.tensor(f["input_ids"],       dtype=torch.long)
                          for f in features], batch_first=True, padding_value=pad_id)
    masks = pad_sequence([torch.tensor(f["attention_mask"],  dtype=torch.long)
                          for f in features], batch_first=True, padding_value=0)
    labs  = pad_sequence([torch.tensor(f["labels"],          dtype=torch.long)
                          for f in features], batch_first=True, padding_value=-100)
    return {"input_ids": ids, "attention_mask": masks, "labels": labs}

# ──────────────────────────────────────────────────────────────────────────
# Label extractor for evaluation
# ──────────────────────────────────────────────────────────────────────────
label_re = re.compile(r"(?i)label\s*:\s*(compliant|non[\s-]?compliant)")

def extract_label(text: str) -> str | None:
    m = label_re.search(text)
    if not m:
        return None
    raw = m.group(1).lower().replace(" ", "-")
    return "Compliant" if raw.startswith("compliant") else "Non-Compliant"

# ──────────────────────────────────────────────────────────────────────────
# Load data & create splits
# ──────────────────────────────────────────────────────────────────────────
df         = pd.read_csv(CSV_PATH)
rows_all   = df.to_dict(orient="records")
labels_all = [r["label"] for r in rows_all]

skf    = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
splits = list(skf.split(np.zeros(len(rows_all)), labels_all))

all_metrics = []

# ──────────────────────────────────────────────────────────────────────────
# Training loop (5‑fold)
# ──────────────────────────────────────────────────────────────────────────
for fold, (train_idx, val_idx) in enumerate(splits, start=1):
    print(f"\n===== Fold {fold}/5 =====")

    train_rows = [rows_all[i] for i in train_idx]
    val_rows   = [rows_all[i] for i in val_idx]

    train_ds = Dataset.from_list([build_sft_example(r) for r in train_rows]).map(
        tokenize_supervised, batched=True, remove_columns=["prompt", "response"]
    )
    val_ds = Dataset.from_list([build_sft_example(r) for r in val_rows]).map(
        tokenize_supervised, batched=True, remove_columns=["prompt", "response"]
    )

    base  = AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float16,
        device_map="auto",
        trust_remote_code=True,
        low_cpu_mem_usage=True,
    )
    model = get_peft_model(base, lora_cfg)

    args = TrainingArguments(
        output_dir=str(OUTPUT_DIR / f"mistral7b_lora_ckpt_fold{fold}"),
        num_train_epochs=3,
        per_device_train_batch_size=2,
        per_device_eval_batch_size=2,
        gradient_accumulation_steps=8,  # effective batch ≈32
        save_steps=250,
        logging_steps=50,
        eval_steps=250,                 # honoured because of eval strategy
        learning_rate=1e-4,
        lr_scheduler_type="cosine",
        optim="adamw_torch",
        fp16=torch.cuda.is_available(),
        report_to="none",
    )
    args.evaluation_strategy = "steps"     # avoids problem on old Transformers

    trainer = Trainer(
        model=model,
        tokenizer=tokenizer,               # fine with deprecation warning
        args=args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        data_collator=causal_lm_collator,  # ← our custom collator
    )
    trainer.train()

    # ── Evaluation ──────────────────────────────────────────────────────
    preds, gold = [], []
    for row in tqdm(val_rows, desc="Evaluating"):
        prompt = build_prompt(row)
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
        with torch.no_grad():
            gen = model.generate(**inputs, max_new_tokens=64, do_sample=False)
        txt   = tokenizer.decode(gen[0], skip_special_tokens=True)
        label = extract_label(txt)
        preds.append(label if label else "UNKNOWN")
        gold.append(row["label"].title())

    valid = {"Compliant", "Non-Compliant"}
    filt_gold, filt_pred = zip(*[(g, p) for g, p in zip(gold, preds)
                                 if g in valid and p in valid])

    acc  = accuracy_score(filt_gold, filt_pred)
    prec = precision_score(filt_gold, filt_pred, pos_label="Compliant")
    rec  = recall_score(filt_gold, filt_pred, pos_label="Compliant")
    f1   = f1_score(filt_gold, filt_pred, pos_label="Compliant")
    cm   = confusion_matrix(filt_gold, filt_pred, labels=["Compliant", "Non-Compliant"])

    print(f"Fold {fold}: acc={acc:.3f}  prec={prec:.3f}  rec={rec:.3f}  f1={f1:.3f}")
    print(cm)

    all_metrics.append({
        "fold": fold, "accuracy": acc, "precision": prec,
        "recall": rec, "f1": f1, "confusion_matrix": cm.tolist()
    })

    # save lightweight adapter and tokenizer
    save_dir = OUTPUT_DIR / f"mistral7b_lora_adapter_fold{fold}"
    model.save_pretrained(save_dir)
    tokenizer.save_pretrained(save_dir)

    # save adapter …
    tokenizer.save_pretrained(save_dir)

    # ── NEW: explicit cleanup ────────────────────────────────
    del trainer, model, base, train_ds, val_ds
    torch.cuda.empty_cache()
    gc.collect()


print("\nCross‑val summary")
print(pd.DataFrame(all_metrics))
```

```python
# mistral_ce_assessor_fixed.py
"""
Fine‑tune Mistral‑7B‑Instruct‑v0.2 with LoRA to classify Cyber‑Essentials answers
───────────────────────────────────────────────────────────────────────────────
Key points
* Supervision format == inference format (two‑line LABEL / REASON)
* Adds BOS token "<s>" expected by Mistral‑Instruct
* Prompt tokens masked with ‑100 in labels
* Robust label extraction via regex
* Custom collator pads *labels* as well as inputs
"""

import gc
import json
import re
from pathlib import Path

import numpy as np
import pandas as pd
import torch
from datasets import Dataset
from peft import LoraConfig, get_peft_model
from sklearn.metrics import (accuracy_score, confusion_matrix, f1_score,
                             precision_score, recall_score)
from sklearn.model_selection import StratifiedKFold
from torch.nn.utils.rnn import pad_sequence
from tqdm.auto import tqdm
from transformers import (AutoModelForCausalLM, AutoTokenizer, Trainer,
                          TrainingArguments, set_seed)

from rubrics import RUBRICS as rubrics      # ← your existing file

# ──────────────────────────────────────────────────────────────────────────
# Config
# ──────────────────────────────────────────────────────────────────────────
SEED = 123
set_seed(SEED)

MODEL_ID   = "mistralai/Mistral-7B-Instruct-v0.2"
CSV_PATH   = "/content/drive/MyDrive/project/CB/balanced_undersampled_per_bucket.csv"
OUTPUT_DIR = Path("/content/drive/MyDrive/project/CB")

lora_cfg = LoraConfig(
    r=16, lora_alpha=32, lora_dropout=0.05, bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True)
tokenizer.pad_token = tokenizer.eos_token

SYSTEM_MSG = (
    "You are a Cyber Essentials assessor.\n"
    "Classify the applicant answer as Compliant or Non-Compliant, using the rubric below.\n"
    "Reply with ONLY TWO lines:\n"
    "LABEL: <Compliant or Non-Compliant>\n"
    "REASON: <Short reason, quoting the rubric criteria met or missed.>"
)

# ──────────────────────────────────────────────────────────────────────────
# Prompt helpers
# ──────────────────────────────────────────────────────────────────────────
def build_prompt(row: dict) -> str:
    rubric = rubrics.get(row["question_id"])
    criteria = "\n".join(rubric["criteria"]) if rubric else "No rubric found."
    user_msg = (
        f"Question:\n{row['question_text']}\n\n"
        f"Applicant answer:\n{row['answer_text']}\n\n"
        f"Rubric:\n{criteria}"
    )
    return (
        "<s>[INST] <<SYS>>\n"
        f"{SYSTEM_MSG}\n"
        "<</SYS>>\n"
        f"{user_msg}\n"
        "[/INST]"
    )

def build_sft_example(row: dict) -> dict:
    prompt   = build_prompt(row)
    response = f"LABEL: {row['label'].title()}\nREASON: {row.get('reason', '')}"
    return {"prompt": prompt, "response": response}

# ──────────────────────────────────────────────────────────────────────────
# Tokenisation
# ──────────────────────────────────────────────────────────────────────────
def tokenize_supervised(batch):
    p_tok = tokenizer(batch["prompt"],   truncation=True, max_length=512, add_special_tokens=False)
    r_tok = tokenizer(batch["response"], truncation=True, max_length=128, add_special_tokens=False)

    input_ids, attention_masks, labels = [], [], []
    for p_ids, r_ids in zip(p_tok["input_ids"], r_tok["input_ids"]):
        ids   = p_ids + r_ids + [tokenizer.eos_token_id]
        mask  = [1] * len(ids)
        lbls  = [-100] * len(p_ids) + r_ids + [tokenizer.eos_token_id]

        input_ids.append(ids)
        attention_masks.append(mask)
        labels.append(lbls)

    return {
        "input_ids": input_ids,
        "attention_mask": attention_masks,
        "labels": labels,
    }

# Custom padding collator that keeps ‑100 for padding in labels
def causal_lm_collator(features):
    pad_id = tokenizer.pad_token_id
    ids   = pad_sequence([torch.tensor(f["input_ids"],       dtype=torch.long)
                          for f in features], batch_first=True, padding_value=pad_id)
    masks = pad_sequence([torch.tensor(f["attention_mask"],  dtype=torch.long)
                          for f in features], batch_first=True, padding_value=0)
    labs  = pad_sequence([torch.tensor(f["labels"],          dtype=torch.long)
                          for f in features], batch_first=True, padding_value=-100)
    return {"input_ids": ids, "attention_mask": masks, "labels": labs}

# ──────────────────────────────────────────────────────────────────────────
# Label extractor for evaluation
# ──────────────────────────────────────────────────────────────────────────
label_re = re.compile(r"(?i)label\s*:\s*(compliant|non[\s-]?compliant)")

def extract_label(text: str) -> str | None:
    m = label_re.search(text)
    if not m:
        return None
    raw = m.group(1).lower().replace(" ", "-")
    return "Compliant" if raw.startswith("compliant") else "Non-Compliant"

# ──────────────────────────────────────────────────────────────────────────
# Load data & create splits
# ──────────────────────────────────────────────────────────────────────────
df         = pd.read_csv(CSV_PATH)
rows_all   = df.to_dict(orient="records")
labels_all = [r["label"] for r in rows_all]

skf    = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
splits = list(skf.split(np.zeros(len(rows_all)), labels_all))

all_metrics = []

# ──────────────────────────────────────────────────────────────────────────
# Training loop (5‑fold)
# ──────────────────────────────────────────────────────────────────────────
for fold, (train_idx, val_idx) in enumerate(splits, start=1):
    print(f"\n===== Fold {fold}/5 =====")

    train_rows = [rows_all[i] for i in train_idx]
    val_rows   = [rows_all[i] for i in val_idx]

    train_ds = Dataset.from_list([build_sft_example(r) for r in train_rows]).map(
        tokenize_supervised, batched=True, remove_columns=["prompt", "response"]
    )
    val_ds = Dataset.from_list([build_sft_example(r) for r in val_rows]).map(
        tokenize_supervised, batched=True, remove_columns=["prompt", "response"]
    )

    base  = AutoModelForCausalLM.from_pretrained(
        MODEL_ID,
        torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float16,
        device_map="auto",
        trust_remote_code=True,
        low_cpu_mem_usage=True,
    )
    model = get_peft_model(base, lora_cfg)

    args = TrainingArguments(
        output_dir=str(OUTPUT_DIR / f"mistral7b_lora_ckpt_fold{fold}"),
        num_train_epochs=3,
        per_device_train_batch_size=2,
        per_device_eval_batch_size=2,
        gradient_accumulation_steps=8,     # effective batch ≈32
        save_steps=250,
        logging_steps=50,
        eval_steps=250,                    # honoured because we set eval strategy below
        learning_rate=1e-4,
        lr_scheduler_type="cosine",
        optim="adamw_torch",
        fp16=torch.cuda.is_available(),
        report_to="none",
    )
    args.evaluation_strategy = "steps"     # avoids problem on old Transformers

    trainer = Trainer(
        model=model,
        tokenizer=tokenizer,               # fine with deprecation warning
        args=args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        data_collator=causal_lm_collator,  # ← our custom collator
    )
    trainer.train()

    # ── Evaluation ──────────────────────────────────────────────────────
    preds, gold = [], []
    for row in tqdm(val_rows, desc="Evaluating"):
        prompt = build_prompt(row)
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
        with torch.no_grad():
            gen = model.generate(**inputs, max_new_tokens=64, do_sample=False)
        txt   = tokenizer.decode(gen[0], skip_special_tokens=True)
        label = extract_label(txt)
        preds.append(label if label else "UNKNOWN")
        gold.append(row["label"].title())

    valid = {"Compliant", "Non-Compliant"}
    filt_gold, filt_pred = zip(*[(g, p) for g, p in zip(gold, preds)
                                 if g in valid and p in valid])

    acc  = accuracy_score(filt_gold, filt_pred)
    prec = precision_score(filt_gold, filt_pred, pos_label="Compliant")
    rec  = recall_score(filt_gold, filt_pred, pos_label="Compliant")
    f1   = f1_score(filt_gold, filt_pred, pos_label="Compliant")
    cm   = confusion_matrix(filt_gold, filt_pred, labels=["Compliant", "Non-Compliant"])

    print(f"Fold {fold}: acc={acc:.3f}  prec={prec:.3f}  rec={rec:.3f}  f1={f1:.3f}")
    print(cm)

    all_metrics.append({
        "fold": fold, "accuracy": acc, "precision": prec,
        "recall": rec, "f1": f1, "confusion_matrix": cm.tolist()
    })

    # save lightweight adapter and tokenizer
    save_dir = OUTPUT_DIR / f"mistral7b_lora_adapter_fold{fold}"
    model.save_pretrained(save_dir)
    tokenizer.save_pretrained(save_dir)

    # save adapter …
    tokenizer.save_pretrained(save_dir)

    # ── NEW: explicit cleanup ────────────────────────────────
    del trainer, model, base, train_ds, val_ds
    torch.cuda.empty_cache()
    gc.collect()


print("\nCross‑val summary")
print(pd.DataFrame(all_metrics))
```

### Train on full dataset

With the best hyperparameters established through cross-validation—namely:




*   num_train_epochs=3,
*   gradient_accumulation_steps=8,
*   per_device_train_batch_size=2,
*   learning_rate=1e-4,
*   fp16=torch.cuda.is_available()
*   lr_scheduler_type="cosine"


which consistently yielded the best results across all folds, I proceeded to train the final model on the full dataset using all available labeled data. At this stage, no portion was held out for validation or testing. This approach maximized the learning signal and enabled the model to fully utilize the dataset in preparation for deployment or inference scenarios.

```python
# mistral_ce_assessor_full_train.py

"""
Fine-tune Mistral-7B-Instruct-v0.2 with LoRA on *all* data (no CV splits)
──────────────────────────────────────────────────────────────────────────────
Key points:
* Supervision format == inference format (two-line LABEL / REASON)
* Adds BOS token "<s>" expected by Mistral-Instruct
* Prompt tokens masked with ‑100 in labels
* Robust label extraction via regex
* Custom collator pads *labels* as well as inputs
* Trains on full data, saves one adapter+tokenizer
"""

import gc
import re
from pathlib import Path

import pandas as pd
import torch
from datasets import Dataset
from peft import LoraConfig, get_peft_model
from torch.nn.utils.rnn import pad_sequence
from transformers import (AutoModelForCausalLM, AutoTokenizer, Trainer,
                          TrainingArguments, set_seed)

from rubrics import RUBRICS as rubrics  # ← your rubric file

# ──────────────────────────────────────
# Config
# ──────────────────────────────────────
SEED = 123
set_seed(SEED)

MODEL_ID   = "mistralai/Mistral-7B-Instruct-v0.2"
CSV_PATH   = "/content/drive/MyDrive/project/CB/balanced_undersampled_per_bucket.csv"
OUTPUT_DIR = Path("/content/drive/MyDrive/project/CB")
FINAL_DIR  = OUTPUT_DIR / "mistral7b_lora_final_full"  # <--- NEW: single final dir

lora_cfg = LoraConfig(
    r=16, lora_alpha=32, lora_dropout=0.05, bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
)

tokenizer = AutoTokenizer.from_pretrained(MODEL_ID, use_fast=True)
tokenizer.pad_token = tokenizer.eos_token

SYSTEM_MSG = (
    "You are a Cyber Essentials assessor.\n"
    "Classify the applicant answer as Compliant or Non-Compliant, using the rubric below.\n"
    "Reply with ONLY TWO lines:\n"
    "LABEL: <Compliant or Non-Compliant>\n"
    "REASON: <Short reason, quoting the rubric criteria met or missed.>"
)

# ─────────────────────────────
# Prompt & tokenization helpers
# ─────────────────────────────
def build_prompt(row: dict) -> str:
    rubric = rubrics.get(row["question_id"])
    criteria = "\n".join(rubric["criteria"]) if rubric else "No rubric found."
    user_msg = (
        f"Question:\n{row['question_text']}\n\n"
        f"Applicant answer:\n{row['answer_text']}\n\n"
        f"Rubric:\n{criteria}"
    )
    return (
        "<s>[INST] <<SYS>>\n"
        f"{SYSTEM_MSG}\n"
        "<</SYS>>\n"
        f"{user_msg}\n"
        "[/INST]"
    )

def build_sft_example(row: dict) -> dict:
    prompt   = build_prompt(row)
    response = f"LABEL: {row['label'].title()}\nREASON: {row.get('reason', '')}"
    return {"prompt": prompt, "response": response}

def tokenize_supervised(batch):
    p_tok = tokenizer(batch["prompt"],   truncation=True, max_length=512, add_special_tokens=False)
    r_tok = tokenizer(batch["response"], truncation=True, max_length=128, add_special_tokens=False)

    input_ids, attention_masks, labels = [], [], []
    for p_ids, r_ids in zip(p_tok["input_ids"], r_tok["input_ids"]):
        ids   = p_ids + r_ids + [tokenizer.eos_token_id]
        mask  = [1] * len(ids)
        lbls  = [-100] * len(p_ids) + r_ids + [tokenizer.eos_token_id]

        input_ids.append(ids)
        attention_masks.append(mask)
        labels.append(lbls)

    return {
        "input_ids": input_ids,
        "attention_mask": attention_masks,
        "labels": labels,
    }

def causal_lm_collator(features):
    pad_id = tokenizer.pad_token_id
    ids   = pad_sequence([torch.tensor(f["input_ids"],       dtype=torch.long)
                          for f in features], batch_first=True, padding_value=pad_id)
    masks = pad_sequence([torch.tensor(f["attention_mask"],  dtype=torch.long)
                          for f in features], batch_first=True, padding_value=0)
    labs  = pad_sequence([torch.tensor(f["labels"],          dtype=torch.long)
                          for f in features], batch_first=True, padding_value=-100)
    return {"input_ids": ids, "attention_mask": masks, "labels": labs}

# ─────────────────────────────
# Load data & train
# ─────────────────────────────
df       = pd.read_csv(CSV_PATH)
rows_all = df.to_dict(orient="records")

full_ds = Dataset.from_list([build_sft_example(r) for r in rows_all]).map(
    tokenize_supervised, batched=True, remove_columns=["prompt", "response"]
)

base  = AutoModelForCausalLM.from_pretrained(
    MODEL_ID,
    torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float16,
    device_map="auto",
    trust_remote_code=True,
    low_cpu_mem_usage=True,
)
model = get_peft_model(base, lora_cfg)

args = TrainingArguments(
    output_dir=str(FINAL_DIR),
    num_train_epochs=3,  # <-- match your CV best or increase for final
    per_device_train_batch_size=2,
    gradient_accumulation_steps=8,
    save_steps=500,   # Less frequent is OK now
    logging_steps=50,
    learning_rate=1e-4,
    lr_scheduler_type="cosine",
    optim="adamw_torch",
    fp16=torch.cuda.is_available(),
    report_to="none",
)
args.evaluation_strategy = "no"   # No evaluation set!

trainer = Trainer(
    model=model,
    tokenizer=tokenizer,
    args=args,
    train_dataset=full_ds,
    data_collator=causal_lm_collator,
)
trainer.train()

# ── Save final adapter and tokenizer ──────────────
model.save_pretrained(FINAL_DIR)
tokenizer.save_pretrained(FINAL_DIR)
print(f"\n✅ Saved final LoRA adapter and tokenizer to {FINAL_DIR}")

# ── Cleanup ──
del trainer, model, base, full_ds
torch.cuda.empty_cache()
gc.collect()
```
