```python
# ── Cell 3 ───────────────────────────────────────────────────────────────
import os, random, time, pandas as pd, spacy
import nlpaug.augmenter.word as naw
from transformers import pipeline
from pathlib import Path
from sentence_transformers import SentenceTransformer, util

# 1) Load CSV
INPUT_PATH = "/content/drive/MyDrive/project/CB/merged_final_combined_corrected.csv"
df = pd.read_csv(INPUT_PATH)

# 2) Build protected-term set
protected = {t.lower() for v in ONTOLOGY_RAW.values() for t in v}
for pat in global_patterns:
    p = pat["pattern"]
    if isinstance(p, str):
        protected.add(p.lower())
    elif isinstance(p, list) and all(isinstance(x, str) for x in p):
        protected.add(" ".join(p).lower())
protected.update(["off","network","firewall","vpn","home","workers","software","service"])

# 3) spaCy + synonym aug
nlp = spacy.load("en_core_web_sm")
syn_aug = naw.SynonymAug(aug_src="wordnet", aug_max=2, stopwords=list(protected))

# 4) MarianMT pipelines on GPU (T4)
bt_en_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr", device=0)
bt_fr_en = pipeline("translation_fr_to_en", model="Helsinki-NLP/opus-mt-fr-en", device=0)

def back_translate_sentencewise_batch(paragraph, batch=32):
    sents = [s.text.strip() for s in nlp(paragraph).sents]
    # → French (batched)
    fr_texts = []
    for i in range(0, len(sents), batch):
        out = bt_en_fr(sents[i:i+batch], max_length=256)
        fr_texts.extend(x["translation_text"] for x in out)
    # → English (batched)
    en_texts = []
    for i in range(0, len(fr_texts), batch):
        out = bt_fr_en(fr_texts[i:i+batch], max_length=256)
        en_texts.extend(x["translation_text"] for x in out)
    return " ".join(en_texts)

# 5) Similarity & word-change guards
st_model = SentenceTransformer("all-MiniLM-L6-v2")

def has_word_change(orig, aug):
    o, a = orig.lower().split(), aug.lower().split()
    diff = sum(1 for x, y in zip(o, a) if x != y) + abs(len(o) - len(a))
    return diff >= 1

# Pre-encode originals once
orig_map = df[~df["file"].str.contains("_aug_")].groupby(
    ["question_id","label"]
)["answer_text"].apply(list).to_dict()
orig_emb_map = {k: st_model.encode(v, convert_to_tensor=True) for k,v in orig_map.items()}

def filter_variants(qid_lbl, orig_text, cands, sim_th=0.90):
    base_embs = orig_emb_map[qid_lbl]
    kept = []
    for c in cands:
        if not has_word_change(orig_text, c): continue
        sim = float(util.cos_sim(st_model.encode(c, convert_to_tensor=True), base_embs).max())
        if sim >= sim_th:
            kept.append(c)
    return kept

def augment_text(qid_lbl, text, do_syn=True, do_bt=True):
    cand_set = set()
    if do_syn:
        out = syn_aug.augment(text)
        cand_set.update(out if isinstance(out, list) else [out])
    if do_bt:
        try:
            cand_set.add(back_translate_sentencewise_batch(text))
        except Exception:
            pass
    cand_set.discard(text)
    return filter_variants(qid_lbl, text, list(cand_set))

# 6) Balance to 100 / (qid,label)
TARGET = 100
meta_cols = [c for c in df.columns if c not in ("file","answer_text")]
aug_rows  = []
start     = time.time()

for (qid, lbl), grp in df.groupby(["question_id","label"]):
    need = TARGET - len(grp)
    if need <= 0: continue

    base_texts = grp["answer_text"].tolist()
    added = 0
    while added < need:
        src        = random.choice(base_texts)
        token_len  = len(src.split())

        # 30 % BT if ≤ 100 tokens
        if token_len <= 100 and random.random() < 0.30:
            variants = augment_text((qid,lbl), src, do_syn=False, do_bt=True)
            method   = "bt"
        else:
            variants = augment_text((qid,lbl), src, do_syn=True,  do_bt=False)
            method   = "syn"

        for v in variants:
            if v.lower() in (t.lower() for t in base_texts): continue
            new_row = {c: grp.iloc[0][c] for c in meta_cols}
            new_row.update({
                "file": f"{Path(INPUT_PATH).stem}_aug_{qid}_{lbl[:3]}_{method}_{added+1}.csv",
                "answer_text": v
            })
            aug_rows.append(new_row)
            base_texts.append(v)
            added += 1
            if added >= need: break

    # progress log
    print(f"{qid}-{lbl}: added {added}/{need}  (elapsed {time.time()-start:.1f}s)")

# 7) Combine & save
df_final = pd.concat([df, pd.DataFrame(aug_rows)], ignore_index=True)
OUT = "/content/drive/MyDrive/project/CB/balanced_100_per_label.csv"
df_final.to_csv(OUT, index=False)
print("Done. Saved to:", OUT)
print("Shape:", df_final.shape, "   Time:", time.time()-start,"s")
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# 1) Load augmented only
aug = pd.read_csv('/content/drive/MyDrive/project/CB/balanced_100_per_label.csv')
aug = aug[aug['file'].str.contains('_aug_')].copy()

model = SentenceTransformer('all-MiniLM-L6-v2')

SIM_THRESHOLD = 0.97      # mark as near-duplicate
near_dupes = []            # (idx_i, idx_j, sim) tuples

for (qid, lbl), grp in aug.groupby(['question_id','label']):
    texts = grp['answer_text'].tolist()
    idxs  = grp.index.tolist()           # keep original dataframe indices
    if len(texts) < 2:
        continue

    # Embed once for the whole bucket
    embs = model.encode(texts, convert_to_tensor=True, batch_size=64)
    # Cosine-similarity matrix (upper-triangle only)
    sims = util.cos_sim(embs, embs).cpu().numpy()

    for i in range(len(texts)):
        for j in range(i+1, len(texts)):
            if sims[i, j] >= SIM_THRESHOLD:
                near_dupes.append(
                    (idxs[i], idxs[j], sims[i, j], qid, lbl)
                )

# 2) Inspect results
dup_df = pd.DataFrame(
    near_dupes,
    columns=['idx_1','idx_2','similarity','question_id','label']
)

print(f"Found {len(dup_df)} augmented pairs with cosine ≥ {SIM_THRESHOLD}")
display(dup_df.head())

# 3) To drop one side of each pair (keep first, drop second):
to_drop = set(dup_df['idx_2'])
clean_df = aug.drop(index=to_drop)

# 4) Combine with originals and save (if you want)
orig = pd.read_csv('/content/drive/MyDrive/project/CB/balanced_100_per_label.csv')
orig = orig[~orig['file'].str.contains('_aug_')]
df_clean = pd.concat([orig, clean_df], ignore_index=True)
df_clean.to_csv('/content/drive/MyDrive/project/CB/balanced_100_per_label_clean.csv',
                index=False)
print("Clean file saved with shape:", df_clean.shape)
```
